```python
#Get the dataset
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv("clean_rainfall.csv",index_col=0)
df.head()
```

```python
#Boxplot
sns.boxplot(x='Humidity', y='ANNUAL', data=df, width=0.5, palette="colorblind")
```

```python
#Scatterplot
sns.regplot(x=df["YEAR"], y=df["DEC"], fit_reg=False)
```

```python
#Histogram
sns.distplot(df['ANNUAL'])
```

```python
#Bar chart
df.iloc[4001,2:14].T.plot(title='2014 Rainfall in Kerala',kind='bar')
```
