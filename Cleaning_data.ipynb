```python
#Get dataset
import pandas as pd
import numpy as np
df = pd.read_csv("dirty_rainfall.csv",index_col=0);
df.head()
```

```python
#Check size of dataset
df.shape
```

```python
#Get info about dataset
df.info()
```

```python
#Check number of missing values
df.isnull().sum()
```

```python
#Clean the numerical values
for col in df.iloc[:,2:15]:
    m = round(df[col].mean(),1)
    df[col].fillna(m, inplace = True)
df.head()
```

```python
#Clean the categorical values
df.fillna(method = 'pad',inplace = True)
df.head()
```

```python
#Check number of missing values
df.isnull().sum()
```

```python
#Load cleaned data into a new file
df.to_csv("clean_rainfall.csv")
```
